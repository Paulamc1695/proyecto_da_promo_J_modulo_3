```python

df = pd.DataFrame(df_info)

# Verificar tipos de valores en el DataFrame
for column in df.columns:
    print(f"Columna: {column}")
    for value in df[column]:
        if pd.isna(value):
            if isinstance(value, float) and np.isnan(value):
                print(f"  {value}: NaN")
            elif value is None:
                print(f"  {value}: None")
            elif pd.isna(value):  # para datetime
                print(f"  {value}: NaT")
            elif isinstance(value, str):
                if value.strip() == '':  # Cadena vacía o solo espacios
                    print('Cadena vacía')

```

```python
#TIPOS DE NULOS 

# 1. (NaN o None):Pandas trata tanto NaN (Not a Number) de NumPy como None de Python como valores nulos. Esto te dará el número total de valores nulos en la columna, pero no distingue entre None y NaN.
                                          #NoneType para valores None y <class 'float'> para NaN (ya que pandas convierte los NaN en float).

# 2. Valores vacios o espacios en blanco: A veces, una columna puede tener cadenas vacías ("") o solo espacios en blanco, que visualmente parecen nulos. 

# 3. Valores especiales como NA o N/A que pueden utilizarse para denotar valores faltantes-
# 4. Valores centinela : como valores negativos -1 o 999
# 5. NaT (Not a Timestamp) para fechas y horas faltantes
```

```python
# Que tipos de nulos tenemos
# 1. Identificar valores nulos (NaN o None): Pandas trata tanto NaN (Not a Number) de NumPy como None de Python como valores nulos. Para verificar la cantidad de nulos en una columna

# Filtra los valores nulos NaN o None y muestra sus tipos
# Filtra las filas donde hay nulos en las columnas especificadas
import numpy as np

# Filtra las filas donde hay nulos en las columnas seleccionadas
columnas = ['BusinessTravel', 'Department', 'EducationField', 'employeenumber',
            'MaritalStatus', 'MonthlyIncome', 'Over18', 'OverTime',
            'PerformanceRating', 'StandardHours', 'TOTALWORKINGYEARS',
            'WORKLIFEBALANCE', 'YearsInCurrentRole', 'SameAsMonthlyIncome',
            'RoleDepartament', 'NUMBERCHILDREN'
]

# Filtra las filas con nulos en las columnas seleccionadas
nulos = df_info[df_info[columnas].isnull().any(axis=1)][columnas]

# Función para identificar si es NaN o None
def tipo_nulo(valor):
    if pd.isna(valor):  # pd.isna() detecta tanto NaN como None
        if isinstance(valor, float) and np.isnan(valor):
            return 'NaN'
        elif valor is None:
            return 'None'
    return 'Otro'

# Aplica la función para determinar el tipo de nulo
tipos_nulos = nulos.applymap(tipo_nulo).apply(pd.Series.unique)

# Imprime los tipos de nulos en cada columna
print(tipos_nulos)


```

```python
# Obtenemos la lista de columnas categóricas que tienen nulos

nulos_columnas_categoricas = df_info[df_info.columns[df_info.isnull().any()]].select_dtypes(include = "O").columns
print("Las columnas categóricas que tienen nulos son : \n ")
print(nulos_columnas_categoricas)
```
